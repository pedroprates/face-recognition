```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
im = cv2.imread('/Users/pedroprates/Google Drive/FaceRecognition/datasets/tcc/0004.jpg')
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(im)
```

```python
os.chdir('/Users/pedroprates/Google Drive/FaceRecognition/src/detect_face')
from face_detector import FaceDetector
```

```python
detector = FaceDetector('/Users/pedroprates/Google Drive/FaceRecognition/models/HaarCascade/haarcascade_frontalface_default.xml')
```

```python
rects = detector.detect_faces(im, minNeighbors=5)
```

```python
faces = detector.extract_faces(im, rects)
```

```python
plt.figure(figsize=(8.0, 6.0))
plt.imshow(faces[0])
plt.grid('off')
```

```python
def save_image(image, name, idx):
    im_to_be_saved = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    cv2.imwrite('/Users/pedroprates/Google Drive/FaceRecognition/datasets/tcc/' + name + '/000' + str(idx) + '.jpg', im_to_be_saved)
```

```python
save_image(faces[0], 'Pedro_Prates', 4)
```
